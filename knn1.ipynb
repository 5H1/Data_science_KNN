```python
#To visualize the decision boundaries of a classifier using the type variable as the target,
#we will need to reduce the dimensionality of the dataset to two dimensions since visualizing in higher dimensions is not feasible. 
#This can be done using dimensionality reduction techniques like PCA (Principal Component Analysis).
#Once we have the two-dimensional data, we can fit a KNN classifier and plot the decision boundaries.
```

```python
#STEP 1 
#Reduce the dataset to two dimensions using PCA.

#STEP 2
#FITTING A MODEL KNN CLASSIFIER TO THE REDUCED DATA
#Fit a KNN classifier to the reduced dataset.

#STEP 3
#isualize the decision boundaries.
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Reduce the dataset to two dimensions using PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# Fit a KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_reduced, y)

# Create a mesh grid for plotting decision boundaries
h = .02  # Step size in the mesh
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict class labels for each point in the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and the original data points
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF', '#AFAFAF', '#FFAAFF', '#AAFFFF', '#FFFFAA'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF', '#7F7F7F', '#FF00FF', '#00FFFF', '#FFFF00'])

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
plt.title("Decision boundaries of KNN classifier")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

```python

#THUS, the above provided is the visualization of boundaries in 2D SPACE.
#This approach is allowing us to see the overall picture of how the KNN Classifier is distinguishing the different classes in the reduced feature space.
```

```python

#if we want to visualize the decision boundaries for every pair of features present in the dataset separately, 
##need to repeat the process for each combination of features. 
```

```python
import itertools

# Plot decision boundaries for every pair of features
features = X.columns
num_features = len(features)

# Create a subplots grid
fig, axes = plt.subplots(num_features, num_features, figsize=(20, 20))

for i, j in itertools.product(range(num_features), range(num_features)):
    if i != j:
        X_pair = X[[features[i], features[j]]]
        X_pair_scaled = scaler.fit_transform(X_pair)
        knn.fit(X_pair_scaled, y)

        x_min, x_max = X_pair_scaled[:, 0].min() - 1, X_pair_scaled[:, 0].max() + 1
        y_min, y_max = X_pair_scaled[:, 1].min() - 1, X_pair_scaled[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

        Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)

        ax = axes[i, j]
        ax.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)
        ax.scatter(X_pair_scaled[:, 0], X_pair_scaled[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
        ax.set_xlabel(features[i])
        ax.set_ylabel(features[j])
    else:
        axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```

```python
#This is showing a grid of plots taht is showing the decision boundaries for every pair of features.
#It  will be going to iterate over all possible combinations of feature pairs,performs scaling and
#train a KNN classifie and then finally plots the decision boundaries.
```

```python
#the below are the key hyperparameters in KNN:

#Number of Neighbors (n_neighbors):

#This is the most important hyperparameter. 
#It will be defining the number of neighbors that are used  for making the prediction. 
#The default value is 5 usually.

#Weights (weights):
#The neighbours contributions are weighted according to this parameter. 

#It can take values:
#'uniform': All neighbors have equal weight (default).
#'distance': Closer neighbors have more weight.

#A user-defined function to compute weights based on distances.

#Algorithm (algorithm):

#this specifies the algorithm used to compute the nearest neighbors. 
#it can iclude the Options as::

#'auto': Lets the algorithm decide the best method (default).

#'ball_tree': it will uses the BallTree algorithm.

#'kd_tree': Uses KDTree algorithm.

#'brute': Uses brute-force search.


#Metric (metric):

#This defines the distance metric to use for the tree. 
#Common choices include:
#'euclidean': Euclidean distance (default).
#'manhattan': Manhattan distance.
#'minkowski': Minkowski distance (generalization of Euclidean and Manhattan distances).

#A user-defined function for distance computation.

#P (p):

#p=1 is equivalent to using the Manhattan distance, 
#while p=2 is equivalent to using the Euclidean distance. 
#The default is 2.

#n_jobs:

#The number of parallel jobs that are to run for neighbors search. 
#-1 means using all processors. The default is 1.
```

```python
#2.What distance metrics can be used in KNN?
```

```python
#In K-Nearest Neighbors (KNN), various distance metrics can be used for measuring the distance between data points. 
#The choice of distance metric can significantly impact the performance of the KNN algorithm.
#the below are some of the common distance metrics:

#Euclidean Distance (metric='euclidean'):

#This is the most commonly used distance metric. 
#It calculates the straight-line distance between two points in Euclidean space.
 
#Manhattan Distance (metric='manhattan'):

#Also known as L1 distance,

#it is going to calculate the distance between two points by summing the absolute differences of their coordinates.

#Minkowski Distance (metric='minkowski'):

#This is a generalization of both Euclidean and Manhattan distances. 
#it uses a parameter k to calculate the distance.
 
#When 𝑝=1
#p=1, it becomes Manhattan distance, and when 

#p=2, it becomes Euclidean distance.
    
#Chebyshev Distance (metric='chebyshev'):

#it is also known as maximum distance and going to calculatethe maximum absolute difference between the coordinates of two points.

#Cosine Distance (metric='cosine'):

#This metric measures the cosine of the angle between two vectors, which is useful for measuring similarity rather than distance.

#Jaccard Distance (metric='jaccard'):

#This is used for comparing the similarity and diversity of sample sets.
```
